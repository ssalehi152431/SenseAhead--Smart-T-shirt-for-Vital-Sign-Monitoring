```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = r"C:\Users\SULTANUS SALEHIN\Downloads\VSM.xlsx"

df = pd.read_excel(file_path, sheet_name="vital_signs")
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.sort_values('timestamp')
```

```python
plt.figure(figsize=(12,5))
sns.histplot(df['tempC'], bins=20, kde=True, color='orange')
plt.title("Temperature Distribution (Consistency Check)")
plt.xlabel("Temperature (C)")
plt.ylabel("Count")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x=df['tempC'], color='orange')
plt.title("Temperature Spread")
plt.xlabel("Temperature (C)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.histplot(df['spo2'].dropna(), bins=10, kde=True, color='green')
plt.title("SpO Distribution (Consistency Check)")
plt.xlabel("SpO (%)")
plt.ylabel("Count")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x=df['spo2'], color='green')
plt.title("SpO Spread")
plt.xlabel("SpO (%)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.histplot(df['motion'], bins=20, kde=True, color='blue')
plt.title("Motion Signal Distribution (Consistency Check)")
plt.xlabel("Motion Index")
plt.ylabel("Count")
plt.grid(True)
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x=df['motion'], color='blue')
plt.title("Motion Index Spread")
plt.xlabel("Motion Index")
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(1,3, figsize=(15,4))
sns.histplot(df['acc_x'], kde=True, ax=axes[0])
sns.histplot(df['acc_y'], kde=True, ax=axes[1])
sns.histplot(df['acc_z'], kde=True, ax=axes[2])

axes[0].set_title("acc_x Dist")
axes[1].set_title("acc_y Dist")
axes[2].set_title("acc_z Dist")

plt.tight_layout()
plt.show()
```
